```python
FEATURE_COLS = data.select_dtypes('float64').columns.tolist() # vou usar somente as colunas float
TARGET = 'history_price'
```

```python
X = data[FEATURE_COLS].drop(columns=['saleEstimate_valueChange.percentageChange', 'history_percentageChange'])
```

```python
y = data[TARGET]
```

```python
print('X (shape): ', X.shape)
print('y (shape): ', y.shape)
```

### Tratanto de outliers (Remoção por quantis)

```python
# ANTES DA REMOÇÃO

q90 = y.quantile(0.90)
q95 = y.quantile(0.95)
q99 = y.quantile(0.99)
q01 = y.quantile(0.01)
q05 = y.quantile(0.05)
q10 = y.quantile(0.10)

ymin = y.min()
ymax = y.max()
ystd = y.std()
median = y.median()
ampl = ymax - ymin

print('=======================================================')
print('====================== QUANTILES ======================')
print('=======================================================\n')
print('90%: ', q90)
print('95%: ', q95)
print('99%: ', q99)
print('01%: ', q01)
print('05%: ', q05)
print('10%: \n\n\n', q10)
print('=======================================================')
print('====================== QUANTILES ======================')
print('=======================================================\n')
print('Min: ', ymin)
print('Max: ', ymax)
print('Std: ', ystd)
print('Median: ', median)
print('Amplitude: ', ampl)
```

```python
# removendo outliers

lower_bound = y.quantile(0.05)  # 5%
upper_bound = y.quantile(0.95)  # 95%

mask = (y >= lower_bound) & (y <= upper_bound)
X = X[mask]
y = y[mask]
```

```python
# após a remoção dos outliers

q90 = y.quantile(0.90)
q95 = y.quantile(0.95)
q99 = y.quantile(0.99)
q01 = y.quantile(0.01)
q05 = y.quantile(0.05)
q10 = y.quantile(0.10)

ymin = y.min()
ymax = y.max()
ystd = y.std()
median = y.median()
ampl = ymax - ymin

print('=======================================================')
print('====================== QUANTILES ======================')
print('=======================================================\n')
print('90%: ', q90)
print('95%: ', q95)
print('99%: ', q99)
print('01%: ', q01)
print('05%: ', q05)
print('10%: \n\n\n', q10)
print('=======================================================')
print('====================== QUANTILES ======================')
print('=======================================================\n')
print('Min: ', ymin)
print('Max: ', ymax)
print('Std: ', ystd)
print('Median: ', median)
print('Amplitude: ', ampl)
```

```python
print('X (shape): ', X.shape)
print('y (shape): ', y.shape)
```

### Tratando de Outliers (Remoção por Z-Score)

```python
from scipy.stats import zscore

y_zscores = zscore(y)
mask = (abs(y_zscores) < 3)  # Mantém valores dentro de 3 desvios padrão
X = X[mask]
y = y[mask]
```

```python
# POR Z-SCORE

q90 = y.quantile(0.90)
q95 = y.quantile(0.95)
q99 = y.quantile(0.99)
q01 = y.quantile(0.01)
q05 = y.quantile(0.05)
q10 = y.quantile(0.10)

ymin = y.min()
ymax = y.max()
ystd = y.std()
median = y.median()
ampl = ymax - ymin

print('=======================================================')
print('====================== QUANTILES ======================')
print('=======================================================\n')
print('90%: ', q90)
print('95%: ', q95)
print('99%: ', q99)
print('01%: ', q01)
print('05%: ', q05)
print('10%: \n\n\n', q10)
print('=======================================================')
print('====================== QUANTILES ======================')
print('=======================================================\n')
print('Min: ', ymin)
print('Max: ', ymax)
print('Std: ', ystd)
print('Median: ', median)
print('Amplitude: ', ampl)
```

```python
print('X (shape): ', X.shape)
print('y (shape): ', y.shape)
```

## Observações
### Antes da Remoção
- Dados extremamente dispersos, com amplitude e desvio padrão elevados.
- Valores extremos afetavam a consistência dos dados.

### Remoção por Quantis
- Menor perda de dados em comparação à abordagem por z-score.
- Dados mais consistentes, com grande redução na amplitude e desvio padrão.
- Preservação da mediana, indicando que os dados centrais não foram muito afetados.

### Remoção por Z-Score
- Menor perda de dados (0,97%) em comparação aos quantis.
- Não é responsável, devido ao assimetria dos dados.

---

## Conclusão
A remoção por quantis apresenta uma menor perda de dados, o que pode ser mais vantajoso dependendo do contexto e da necessidade de maior volume de dados para o modelo.
